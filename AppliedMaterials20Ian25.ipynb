1. **Rentabilitatea activelor (ROA)** – Return on Assets.
2. **Rentabilitatea capitalurilor proprii (ROE)** – Return on Equity.
3. **Marja brută** – Gross Margin.
4. **Marja netă** – Net Profit Margin.
5. **Rata curentă** – Current Ratio (Active curente / Datorii curente).

### Cod Python pentru analiza indicatorilor financiari:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Citim datele financiare dintr-un fișier CSV (datele preluate anterior)
file_path = 'AMAT_financial_data.csv'  # Înlocuiește cu calea către fișierul tău CSV
financial_data = pd.read_csv(file_path)

# Afișăm primele câteva rânduri pentru a verifica datele
print(financial_data.head())

# 1. Calcularea ROA (Return on Assets) = Net Income / Total Assets
financial_data['ROA (%)'] = (financial_data['Net Income (USD)'] / financial_data['Total Assets (USD)']) * 100

# 2. Calcularea ROE (Return on Equity) = Net Income / Total Shareholder Equity
financial_data['ROE (%)'] = (financial_data['Net Income (USD)'] / financial_data['Total Shareholder Equity (USD)']) * 100

# 3. Calcularea Gross Margin (Marja Bruta) = Gross Profit / Total Revenue
financial_data['Gross Margin (%)'] = (financial_data['Gross Profit (USD)'] / financial_data['Total Revenue (USD)']) * 100

# 4. Calcularea Net Profit Margin (Marja Netă) = Net Income / Total Revenue
financial_data['Net Profit Margin (%)'] = (financial_data['Net Income (USD)'] / financial_data['Total Revenue (USD)']) * 100

# 5. Calcularea Rata curentă (Current Ratio) = Total Assets / Total Liabilities
financial_data['Current Ratio'] = financial_data['Total Assets (USD)'] / financial_data['Total Liabilities (USD)']

# Afișăm tabelul cu noii indicatori calculați
print("\nIndicatori financiari calculați:")
print(financial_data[['Fiscal Year', 'ROA (%)', 'ROE (%)', 'Gross Margin (%)', 'Net Profit Margin (%)', 'Current Ratio']].head())

# Vizualizarea evoluției ROA și ROE în timp
plt.figure(figsize=(10, 6))
plt.plot(financial_data['Fiscal Year'], financial_data['ROA (%)'], marker='o', label='ROA (%)', color='blue')
plt.plot(financial_data['Fiscal Year'], financial_data['ROE (%)'], marker='o', label='ROE (%)', color='green')

plt.show()

plt.plot(financial_data['Fiscal Year'], financial_data['Net Profit Margin (%)'], marker='o', label='Net Profit Margin (%)', color='red')
plt.title('Evoluția Marjelor Brute și Nete în timp')

plt.show()
```

### Explicații:

1. **Rentabilitatea activelor (ROA)**: Măsoară eficiența companiei în utilizarea activelor pentru a genera profituri.
   - Formula: `ROA = Net Income / Total Assets`.
   
2. **Rentabilitatea capitalurilor proprii (ROE)**: Măsoară cât de eficient utilizează compania capitalul propriu pentru a genera profit.
   - Formula: `ROE = Net Income / Total Shareholder Equity`.

3. **Marja brută (Gross Margin)**: Măsoară procentul veniturilor rămas după acoperirea costurilor directe de producție.
   - Formula: `Gross Margin = Gross Profit / Total Revenue`.

4. **Marja netă (Net Profit Margin)**: Arată procentul veniturilor totale care rămâne sub formă de profit net după toate cheltuielile.
   - Formula: `Net Profit Margin = Net Income / Total Revenue`.

5. **Rata curentă (Current Ratio)**: Evaluează capacitatea companiei de a-și plăti datoriile pe termen scurt.
   - Formula: `Current Ratio = Total Assets / Total Liabilities`.

````python
import pandas as pd
import numpy as np

# Citim datele financiare dintr-un fișier CSV (preluate anterior)
file_path = 'AMAT_financial_data.csv'  # Înlocuiește cu calea către fișierul tău CSV
financial_data = pd.read_csv(file_path)

# Calculăm rata anuală compusă de creștere (CAGR) pentru veniturile companiei pe o perioadă definită
start_year = financial_data['Fiscal Year'].min()
end_year = financial_data['Fiscal Year'].max()

revenue_start = financial_data[financial_data['Fiscal Year'] == start_year]['Total Revenue (USD)'].values[0]
revenue_end = financial_data[financial_data['Fiscal Year'] == end_year]['Total Revenue (USD)'].values[0]

years = end_year - start_year
cagr_revenue = (revenue_end / revenue_start) ** (1 / years) - 1

print(f"CAGR pentru veniturile totale între {start_year} și {end_year} este: {cagr_revenue:.2%}")

# Proiecție pentru următorii 5 ani, folosind CAGR calculat
future_years = 5
future_revenues = [revenue_end * (1 + cagr_revenue) ** i for i in range(1, future_years + 1)]

# Afișăm proiecțiile
print("\nProiecții pentru veniturile viitoare:")
for i, revenue in enumerate(future_revenues, 1):
    print(f"Anul {end_year + i}: Venituri proiectate: {revenue:,.2f} USD")
````

Pentru a proiecta prețul unei acțiuni în funcție de creșterea veniturilor, putem folosi o abordare simplă bazată pe multipli financiari, cum ar fi raportul **P/S (Price-to-Sales)**. Acest raport compară prețul unei acțiuni cu veniturile generate de companie pe acțiune. O formulă simplificată pentru proiecția prețului acțiunii în viitor poate fi:

\[
\text{Proiecția prețului acțiunii} = \text{P/S Ratio} \times \text{Venituri totale pe acțiune}
\]

Pașii pentru a face această proiecție sunt:

1. **Calculează veniturile pe acțiune** pentru perioada actuală și pentru perioada viitoare.
2. **Stabilește un P/S Ratio**. Acest indicator poate rămâne constant în proiecția noastră sau poate varia în funcție de condițiile pieței și evaluarea companiei.
3. **Proiectează prețul acțiunii** pe baza creșterii veniturilor și a P/S Ratio.

### Pasul 1: Calculul Veniturilor pe Acțiune
Veniturile pe acțiune se calculează astfel:

\[
\text{Venituri pe acțiune} = \frac{\text{Venituri totale}}{\text{Numărul de acțiuni în circulație}}
\]

### Pasul 2: P/S Ratio
Raportul P/S variază de la o companie la alta și poate fi calculat astfel:

\[
\text{P/S Ratio} = \frac{\text{Prețul acțiunii curente}}{\text{Venituri pe acțiune curente}}
\]

### Pasul 3: Proiectarea prețului acțiunii
Folosim veniturile pe acțiune viitoare și P/S Ratio pentru a calcula prețul acțiunii proiectat:

\[
\text{Prețul acțiunii proiectat} = \text{P/S Ratio} \times \text{Venituri pe acțiune viitoare}
\]

### Cod Python pentru proiecția prețului acțiunii

```python
import pandas as pd

# Date de intrare - date financiare și prețul actual al acțiunii
file_path = 'AMAT_financial_data.csv'  # Înlocuiește cu fișierul tău de date
financial_data = pd.read_csv(file_path)

# Date curente despre prețul acțiunii și numărul de acțiuni în circulație (acestea trebuie să fie disponibile)
current_stock_price = 140  # Înlocuiește cu prețul curent al acțiunii AMAT
shares_outstanding = 900_000_000  # Înlocuiește cu numărul de acțiuni în circulație

# Calculăm veniturile totale curente și pe acțiune
latest_year = financial_data['Fiscal Year'].max()
total_revenue = financial_data[financial_data['Fiscal Year'] == latest_year]['Total Revenue (USD)'].values[0]
revenue_per_share = total_revenue / shares_outstanding

# Calculăm P/S Ratio curent
ps_ratio = current_stock_price / revenue_per_share

# Proiecții de venituri calculate anterior pentru următorii 5 ani
future_revenues = [total_revenue * 1.07**i for i in range(1, 6)]  # Presupunem o creștere anuală de 7%

# Proiectăm veniturile pe acțiune și prețul acțiunii pentru următorii 5 ani
future_revenue_per_share = [revenue / shares_outstanding for revenue in future_revenues]
projected_stock_prices = [ps_ratio * revenue_per_share for revenue_per_share in future_revenue_per_share]

# Afișăm proiecțiile de preț ale acțiunii
print(f"P/S Ratio curent: {ps_ratio:.2f}")
print("\nProiecțiile prețului acțiunii pentru următorii 5 ani:")
for i, price in enumerate(projected_stock_prices, 1):
    print(f"Anul {latest_year + i}: Preț proiectat al acțiunii: ${price:.2f}")
```

### Explicații:
1. **Date de intrare**:
   - Prețul acțiunii curente (`current_stock_price`): Acesta este prețul actual al acțiunii **AMAT** pe piață.
   - Numărul de acțiuni în circulație (`shares_outstanding`): Este totalul acțiunilor disponibile pe piață.
   
2. **Calcule**:
   - Calculăm **veniturile pe acțiune** pentru anul curent, împărțind veniturile totale la numărul de acțiuni.
   - Calculăm **P/S Ratio**, un raport care leagă prețul curent al acțiunii de veniturile curente pe acțiune.
   - Proiectăm veniturile viitoare pe acțiune, folosind o rată de creștere a veniturilor (7% în acest exemplu).
   
3. **Proiecțiile prețului acțiunii**:
   - Pentru următorii 5 ani, calculăm prețul proiectat al acțiunii bazat pe veniturile viitoare și pe P/S Ratio curent.

### Concluzie:
- **P/S Ratio** ne permite să proiectăm prețul acțiunii în funcție de creșterea veniturilor.
- Proiecțiile oferite sunt bazate pe presupunerea că **P/S Ratio** rămâne constant. În realitate, acest raport poate varia în funcție de evoluția pieței și de percepția investitorilor asupra companiei.
  
Aceasta este o metodă simplificată și ar trebui combinată cu alți factori de evaluare, cum ar fi profitabilitatea, fluxurile de numerar și starea pieței, pentru o imagine completă asupra potențialului de creștere al prețului acțiunii.

````python
import pandas as pd

# Date de intrare - date financiare și prețul actual al acțiunii
file_path = 'AMAT_financial_data.csv'  # Înlocuiește cu fișierul tău de date
financial_data = pd.read_csv(file_path)

# Date curente despre prețul acțiunii și numărul de acțiuni în circulație (acestea trebuie să fie disponibile)
current_stock_price = 140  # Înlocuiește cu prețul curent al acțiunii AMAT
shares_outstanding = 900_000_000  # Înlocuiește cu numărul de acțiuni în circulație

# Calculăm veniturile totale curente și pe acțiune
latest_year = financial_data['Fiscal Year'].max()
total_revenue = financial_data[financial_data['Fiscal Year'] == latest_year]['Total Revenue (USD)'].values[0]
revenue_per_share = total_revenue / shares_outstanding

# Calculăm P/S Ratio curent
ps_ratio = current_stock_price / revenue_per_share

# Proiecții de venituri calculate anterior pentru următorii 5 ani
future_revenues = [total_revenue * 1.07**i for i in range(1, 6)]  # Presupunem o creștere anuală de 7%

# Proiectăm veniturile pe acțiune și prețul acțiunii pentru următorii 5 ani
future_revenue_per_share = [revenue / shares_outstanding for revenue in future_revenues]
projected_stock_prices = [ps_ratio * revenue_per_share for revenue_per_share in future_revenue_per_share]

# Afișăm proiecțiile de preț ale acțiunii
print(f"P/S Ratio curent: {ps_ratio:.2f}")
print("\nProiecțiile prețului acțiunii pentru următorii 5 ani:")
for i, price in enumerate(projected_stock_prices, 1):
    print(f"Anul {latest_year + i}: Preț proiectat al acțiunii: ${price:.2f}")
````

În acest scop, vom folosi **medii mobile**, analiza pantei și alți indicatori tehnici pentru a identifica diverse trenduri. 

### Metode pentru identificarea trendurilor:
1. **Medii mobile simple (SMA)**: Calcularea mediei mobile pe un număr de zile poate arăta direcția trendului. Dacă prețul se află deasupra mediei mobile și SMA este în creștere, avem un **trend ascendent**. Dacă prețul este sub SMA și aceasta scade, avem un **trend descendent**.
   
2. **Analiza pantei**: Calcularea pantei între două puncte consecutive ale prețului sau ale mediei mobile poate indica trendul. O pantă pozitivă sugerează un **trend ascendent**, o pantă negativă sugerează un **trend descendent**, iar o pantă aproape de zero poate sugera un **trend lateral**.

3. **Indicatorul RSI (Relative Strength Index)**: Acesta poate fi folosit pentru a evalua dacă o acțiune este într-o zonă de **supracumpărare** (peste 70) sau **supravânzare** (sub 30), indicând potențialele schimbări de trend.

plt.plot(stock_data['Date'], stock_data['SMA_200'], label='SMA 200 zile', color='red')

plt.show()
