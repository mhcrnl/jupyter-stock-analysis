```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download GE stock data
ge = yf.download('GE', start='2021-01-01', end='2023-06-18')

# Calculate the moving average
window_size = 50
ge['Moving Average'] = ge['Close'].rolling(window=window_size).mean()

# Calculate upper and lower bands
envelope_percentage = 5  # Adjust this value to set the percentage for the envelope
ge['Upper Band'] = ge['Moving Average'] * (1 + envelope_percentage / 100)
ge['Lower Band'] = ge['Moving Average'] * (1 - envelope_percentage / 100)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(ge.index, ge['Close'], label='GE Stock')
plt.plot(ge.index, ge['Moving Average'], label='Moving Average ({} days)'.format(window_size))
plt.plot(ge.index, ge['Upper Band'], 'r--', label='Upper Band ({}% envelope)'.format(envelope_percentage))
plt.plot(ge.index, ge['Lower Band'], 'g--', label='Lower Band ({}% envelope)'.format(envelope_percentage))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GE Stock Data with Moving Average Envelope')
plt.legend()
plt.grid(True)
plt.show()
```

### Moving Average Deviation
***
Moving average deviation, also known as the difference between the closing price and the moving average, measures the distance or deviation of a data point (in this case, the closing price) from the calculated moving average.

To calculate the moving average deviation, you subtract the moving average value from the closing price for each data point. The result gives you a measure of how far each data point deviates from the moving average.

This deviation can provide insights into the volatility or trend strength of a dataset. Larger deviations indicate greater price fluctuations and potential trading opportunities, while smaller deviations suggest a more stable price pattern.

In financial analysis and technical analysis, moving average deviation can be used to identify potential buy or sell signals. Traders may look for instances where the price moves significantly above or below the moving average, indicating potential overbought or oversold conditions.

By incorporating moving average deviation into your analysis, you can gain a deeper understanding of the price movements and potential trends in a dataset.
***

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download GE stock data
ge = yf.download('GE', start='2021-01-01', end='2023-06-18')

# Calculate the moving average
window_size = 50
ge['Moving Average'] = ge['Close'].rolling(window=window_size).mean()

# Calculate the moving average deviation
ge['Moving Average Deviation'] = ge['Close'] - ge['Moving Average']

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(ge.index, ge['Close'], label='GE Stock')
plt.plot(ge.index, ge['Moving Average'], label='Moving Average (50 days)')
plt.plot(ge.index, ge['Moving Average Deviation'], label='Moving Average Deviation')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GE Stock Data with Moving Average Deviation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime

# Download GE stock data
ge = yf.download('GE', start='2021-01-01', end=datetime.datetime.now())

# Calculate the moving average
window_size = 20
ge['Moving Average'] = ge['Close'].rolling(window=window_size).mean()

# Calculate the standard deviation
ge['Std Deviation'] = ge['Close'].rolling(window=window_size).std()

# Calculate the upper and lower Bollinger Bands
ge['Upper Band'] = ge['Moving Average'] + (2 * ge['Std Deviation'])
ge['Lower Band'] = ge['Moving Average'] - (2 * ge['Std Deviation'])

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(ge.index, ge['Close'], label='GE Stock')
plt.plot(ge.index, ge['Moving Average'], label='Moving Average (20 days)')
plt.plot(ge.index, ge['Upper Band'], label='Upper Band')
plt.plot(ge.index, ge['Lower Band'], label='Lower Band')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GE Stock Data with Bollinger Bands')
plt.legend()
plt.grid(True)
plt.show()
```

The relative strength index (RSI) is a momentum indicator used in technical analysis. RSI measures the speed and magnitude of a security's recent price changes to evaluate overvalued or undervalued conditions in the price of that security.

The RSI is displayed as an oscillator (a line graph) on a scale of zero to 100. The indicator was developed by J. Welles Wilder Jr. and introduced in his seminal 1978 book, New Concepts in Technical Trading Systems.
1
J. Welles Wilder Jr. "New Concepts in Technical Trading Systems." Trend Research, 1978.


The RSI can do more than point to overbought and oversold securities. It can also indicate securities that may be primed for a trend reversal or corrective pullback in price. It can signal when to buy and sell. Traditionally, an RSI reading of 70 or above indicates an overbought situation. A reading of 30 or below indicates an oversold condition.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download GE stock data
ge = yf.download('GE', start='2021-01-01', end='2023-06-18')

# Calculate the RSI
window_size = 14
ge['Change'] = ge['Close'].diff()
ge['Gain'] = ge['Change'].apply(lambda x: x if x > 0 else 0)
ge['Loss'] = ge['Change'].apply(lambda x: abs(x) if x < 0 else 0)
ge['Avg Gain'] = ge['Gain'].rolling(window=window_size).mean()
ge['Avg Loss'] = ge['Loss'].rolling(window=window_size).mean()
ge['RS'] = ge['Avg Gain'] / ge['Avg Loss']
ge['RSI'] = 100 - (100 / (1 + ge['RS']))

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(ge.index, ge['Close'], label='GE Stock')
plt.plot(ge.index, ge['RSI'], label='RSI (14 days)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GE Stock Data with RSI')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime

# Download GE stock data
ge = yf.download('GE', start='2021-01-01', end=datetime.datetime.today())

# Calculate MACD
ema_12 = ge['Close'].ewm(span=12, adjust=False).mean()
ema_26 = ge['Close'].ewm(span=26, adjust=False).mean()
macd = ema_12 - ema_26
signal_line = macd.ewm(span=9, adjust=False).mean()
histogram = macd - signal_line

# Plot the data
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(ge.index, ge['Close'], label='GE Stock')
plt.ylabel('Price')
plt.title('GE Stock Data')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(ge.index, macd, label='MACD', color='blue')
plt.plot(ge.index, signal_line, label='Signal Line', color='red')
plt.bar(ge.index, histogram, label='Histogram', color='gray')
plt.xlabel('Date')
plt.ylabel('MACD')
plt.title('MACD Indicator')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

Next, we calculate the Accumulation/Distribution Line (ADL) using the following formula:

ADL = ((Close - Low) - (High - Close)) / (High - Low) * Volume

We calculate the difference between the closing price and the range between the high and low prices, divide it by the price range, and multiply it by the volume. We then calculate the cumulative sum of the ADL values.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download GE stock data
ge = yf.download('GE', start='2021-01-01', end='2023-06-18')

# Calculate the Accumulation/Distribution Line (ADL)
ge['ADL'] = ((ge['Close'] - ge['Low']) - (ge['High'] - ge['Close'])) / (ge['High'] - ge['Low'])
ge['ADL'] = ge['ADL'] * ge['Volume']
ge['ADL'] = ge['ADL'].cumsum()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(ge.index, ge['Close'], label='GE Stock')
plt.plot(ge.index, ge['ADL'], label='Accumulation/Distribution Line (ADL)')
plt.xlabel('Date')
plt.ylabel('Price/ADL')
plt.title('GE Stock Data with Accumulation/Distribution Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
from fbprophet import Prophet

# Download GM stock data
gm = yf.download('GM', start='2010-01-01', end='2023-06-18')
gm = gm[['Close']].reset_index()
gm = gm.rename(columns={'Date': 'ds', 'Close': 'y'})

# Download GRAAF stock data
graaf = yf.download('GRAAF', start='2010-01-01', end='2023-06-18')
graaf = graaf[['Close']].reset_index()
graaf = graaf.rename(columns={'Date': 'ds', 'Close': 'y'})

# Create and fit the Prophet models
gm_model = Prophet(daily_seasonality=True)
gm_model.fit(gm)

graaf_model = Prophet(daily_seasonality=True)
graaf_model.fit(graaf)

# Predict future prices
gm_future = gm_model.make_future_dataframe(periods=365)  # Forecasting for the next 365 days
gm_forecast = gm_model.predict(gm_future)

graaf_future = graaf_model.make_future_dataframe(periods=365)  # Forecasting for the next 365 days
graaf_forecast = graaf_model.predict(graaf_future)

# Plot the forecasts
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

ax[0].plot(gm['ds'], gm['y'], label='GM Stock')
ax[0].plot(gm_forecast['ds'], gm_forecast['yhat'], label='GM Forecast')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Price')
ax[0].set_title('GM Stock Price Forecast')

ax[1].plot(graaf['ds'], graaf['y'], label='GRAAF Stock')
ax[1].plot(graaf_forecast['ds'], graaf_forecast['yhat'], label='GRAAF Forecast')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Price')
ax[1].set_title('GRAAF Stock Price Forecast')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Download GM stock data
gm = yf.download('GM', start='2015-01-01', end='2023-06-18')

# Extract the 'Close' prices
gm = gm[['Close']].reset_index()

# Create a new column 'Days' representing the number of days from the start date
gm['Days'] = (gm['Date'] - gm['Date'].min()).dt.days

# Train a linear regression model
model = LinearRegression()
model.fit(gm[['Days']], gm['Close'])

# Predict future prices
future_days = pd.DataFrame({'Days': range(gm['Days'].max() + 1, gm['Days'].max() + 31)})
future_prices = model.predict(future_days)

# Create a DataFrame for future dates and prices
future_dates = pd.date_range(start=gm['Date'].max() + pd.DateOffset(days=1), periods=30, freq='D')
future_data = pd.DataFrame({'Date': future_dates, 'Close': future_prices})

# Concatenate the historical and future data
gm_forecast = pd.concat([gm, future_data], ignore_index=True)

# Plot the historical and predicted prices
plt.figure(figsize=(12, 6))
plt.plot(gm_forecast['Date'], gm_forecast['Close'], label='GM Stock')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GM Stock Price Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import datetime

# Download GM stock data
gm = yf.download('GM', start='2022-01-01', end=datetime.datetime.today())

# Prepare the data
gm['Date'] = gm.index
gm['Days'] = (gm['Date'] - gm['Date'].min()).dt.days
gm['Price'] = gm['Close']

# Train the linear regression model
model = LinearRegression()
model.fit(gm[['Days']], gm[['Price']])

# Predict the future prices
future_days = np.arange(gm['Days'].max(), gm['Days'].max() + 100)
future_prices = model.predict(future_days.reshape(-1, 1))

# Plot the data and predictions
plt.figure(figsize=(12, 6))
plt.scatter(gm['Days'], gm['Price'], color='blue', label='Historical Price')
plt.plot(future_days, future_prices, color='red', label='Predicted Price')
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('GM Stock Price Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
