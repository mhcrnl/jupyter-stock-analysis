# 2. Ford (F)

```python
simbol = "F"
start  = dt.datetime(2000,1,1)
end    = dt.datetime.now()

f = web.get_data_yahoo(simbol,start,end)

mpl.plot(f, 
         type='line',
         title = "Ford 20 ani pret.",
         ylabel = "Pret",
         xlabel = "Timp",
         #mav=(3,6,9)
         figsize=(20,12)
        )

mpl.__version__
```

```python
f30 = f.tail(30)
mpl.plot(f30, 
         type='candle',
         title = "Ford 30 zile",
         volume=True,
         figsize=(20,10)
        )
```

## 3. IVR

```python
simbol = "IVR"
start  = dt.datetime(2000,1,1)
end    = dt.datetime.now()

ivr = web.get_data_yahoo(simbol,start,end)

mpl.plot(ivr, 
         type='line',
         title = "IVR pret.",
         ylabel = "Pret",
         xlabel = "Timp",
         #mav=(3,6,9)
         figsize=(20,12)
        )

mpl.__version__
```

```python
ivr30 = ivr.tail(30)
mpl.plot(ivr30, 
         type='candle',
         title = "IVR 30 zile",
         volume=True,
         figsize=(20,10)
        )
```
