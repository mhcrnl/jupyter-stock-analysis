```python
# Instalare dependente
#!pip install yfinance matplotlib fpdf

# Importăm librăriile necesare
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from fpdf import FPDF

# Definim clasa pentru analiza acțiunii
class StockAnalysis:
    def __init__(self, symbol, start_date, end_date):
        self.symbol = symbol
        self.start_date = start_date
        self.end_date = end_date
        self.data = self.get_stock_data()

    # Funcție pentru obținerea datelor istorice
    def get_stock_data(self):
        data = yf.download(self.symbol, start=self.start_date, end=self.end_date)
        return data
    
    # Funcție pentru calcularea indicatorilor tehnici
    def calculate_technical_indicators(self):
        data = self.data
        # SMA - Simple Moving Average
        data['SMA_50'] = data['Close'].rolling(window=50).mean()
        data['SMA_200'] = data['Close'].rolling(window=200).mean()

        # RSI - Relative Strength Index
        delta = data['Close'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
        rs = gain / loss
        data['RSI'] = 100 - (100 / (1 + rs))

        # MACD - Moving Average Convergence Divergence
        fast_ema = data['Close'].ewm(span=12, adjust=False).mean()
        slow_ema = data['Close'].ewm(span=26, adjust=False).mean()
        data['MACD'] = fast_ema - slow_ema
        data['MACD_signal'] = data['MACD'].ewm(span=9, adjust=False).mean()

        # Stochastic Oscillator
        low_min = data['Low'].rolling(window=14).min()
        high_max = data['High'].rolling(window=14).max()
        data['Stochastic'] = (data['Close'] - low_min) / (high_max - low_min) * 100

        # Bollinger Bands
        data['Bollinger_Mid'] = data['Close'].rolling(window=20).mean()
        data['Bollinger_Upper'] = data['Bollinger_Mid'] + 2 * data['Close'].rolling(window=20).std()
        data['Bollinger_Lower'] = data['Bollinger_Mid'] - 2 * data['Close'].rolling(window=20).std()

        # Parabolic SAR
        data['SAR'] = self.calculate_parabolic_sar(data)

        self.data = data
    
    # Funcție pentru calcularea Parabolic SAR
    def calculate_parabolic_sar(self, data, acceleration=0.02, max_acceleration=0.2):
        sar = data['Close'].copy()
        ep = data['High'][0]
        af = acceleration
        trend = 1
        for i in range(1, len(data)):
            if trend == 1:
                sar[i] = sar[i - 1] + af * (ep - sar[i - 1])
                if data['Low'][i] < sar[i]:
                    trend = -1
                    sar[i] = ep
                    ep = data['Low'][i]
                    af = acceleration
            else:
                sar[i] = sar[i - 1] + af * (ep - sar[i - 1])
                if data['High'][i] > sar[i]:
                    trend = 1
                    sar[i] = ep
                    ep = data['High'][i]
                    af = acceleration
            if af < max_acceleration:
                af += acceleration
        return sar

    # Funcție pentru generarea graficelor
    def generate_charts(self):
        data = self.data
        plt.figure(figsize=(10, 12))

        # Grafic pentru prețul acțiunii și SMAs
        plt.subplot(3, 2, 1)
        plt.plot(data['Close'], label='Close Price', color='black')
        plt.plot(data['SMA_50'], label='50-day SMA', color='blue')
        plt.plot(data['SMA_200'], label='200-day SMA', color='red')
        plt.title('Stock Price with SMAs')
        plt.legend()

        # Grafic pentru RSI
        plt.subplot(3, 2, 2)
        plt.plot(data['RSI'], label='RSI', color='green')
        plt.axhline(y=70, color='r', linestyle='--')
        plt.axhline(y=30, color='r', linestyle='--')
        plt.title('Relative Strength Index (RSI)')
        plt.legend()

        # Grafic pentru MACD
        plt.subplot(3, 2, 3)
        plt.plot(data['MACD'], label='MACD', color='blue')
        plt.plot(data['MACD_signal'], label='MACD Signal', color='red')
        plt.title('MACD')
        plt.legend()

        # Grafic pentru Stochastic Oscillator
        plt.subplot(3, 2, 4)
        plt.plot(data['Stochastic'], label='Stochastic Oscillator', color='purple')
        plt.axhline(y=80, color='r', linestyle='--')
        plt.axhline(y=20, color='r', linestyle='--')
        plt.title('Stochastic Oscillator')
        plt.legend()

        # Grafic pentru Bollinger Bands
        plt.subplot(3, 2, 5)
        plt.plot(data['Close'], label='Close Price', color='black')
        plt.plot(data['Bollinger_Upper'], label='Upper Bollinger Band', color='red')
        plt.plot(data['Bollinger_Lower'], label='Lower Bollinger Band', color='green')
        plt.title('Bollinger Bands')
        plt.legend()

        # Grafic pentru Parabolic SAR
        plt.subplot(3, 2, 6)
        plt.plot(data['Close'], label='Close Price', color='black')
        plt.plot(data['SAR'], label='Parabolic SAR', color='orange')
        plt.title('Parabolic SAR')
        plt.legend()

        plt.tight_layout()
        plt.savefig('technical_indicators.png')

    # Funcție pentru crearea unui raport PDF
    def create_pdf_report(self, filename='stock_report.pdf'):
        pdf = FPDF()
        pdf.add_page()
        
        # Titlu raport
        pdf.set_font('Arial', 'B', 16)
        pdf.cell(200, 10, f'{self.symbol} Stock Analysis Report', ln=True, align='C')
        
        # Detalii acțiune
        pdf.set_font('Arial', '', 12)
        pdf.ln(10)
        pdf.cell(200, 10, f'Stock Symbol: {self.symbol}', ln=True)
        pdf.cell(200, 10, f'Analysis period: {self.start_date} to {self.end_date}', ln=True)
        
        # Inserăm graficul
        pdf.ln(10)
        pdf.image('technical_indicators.png', x=10, y=50, w=180)
        
        # Indicatori tehnici
        pdf.ln(120)
        pdf.cell(200, 10, 'Technical Indicators Summary:', ln=True)
        pdf.cell(200, 10, f'Last Close Price: {self.data["Close"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'50-day SMA: {self.data["SMA_50"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'200-day SMA: {self.data["SMA_200"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'RSI: {self.data["RSI"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'MACD: {self.data["MACD"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'MACD Signal: {self.data["MACD_signal"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'Stochastic: {self.data["Stochastic"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'Bollinger Upper: {self.data["Bollinger_Upper"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'Bollinger Lower: {self.data["Bollinger_Lower"].iloc[-1]:.2f}', ln=True)
        pdf.cell(200, 10, f'Parabolic SAR: {self.data["SAR"].iloc[-1]:.2f}', ln=True)

        # Salvăm raportul PDF
        pdf.output(filename)

# Creăm obiectul pentru analiza acțiunii Palantir
analysis = StockAnalysis('PLTR', '2022-01-01', '2025-01-01')

# Calculăm indicatorii tehnici
analysis.calculate_technical_indicators()

# Generăm graficele
analysis.generate_charts()

# Creăm raportul PDF
analysis.create_pdf_report()
```
