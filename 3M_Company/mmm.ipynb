```python
import pandas as pd
import yfinance as yf

# Simbolul acțiunii pe care dorești să o analizezi
symbol = "MMM"  # Exemplu: Apple Inc. (AAPL)

# Descarcarea datelor de preț pentru intervalul de 52 de săptămâni
stock = yf.Ticker(symbol)
data = stock.history(period="1y")

# Crearea unui DataFrame cu datele
df = pd.DataFrame(data)

# Afișarea tabelului cu datele de preț
table = df[["Low", "High"]].describe().transpose()
table = table.round(2)

print(table)
```
